```python
crema = "/kaggle/input/cremad/AudioWAV/"
crema_directory_list = os.listdir(crema)
file_name = []
file_emotion = []
file_path = []
file_id = []

for file in crema_directory_list:
    # storing file paths
    file_name.append(file.split('.')[0])
    file_path.append(crema + file)
    # storing file emotions
    part=file.split('_')
    if part[2] == 'SAD':
        file_emotion.append('Sadness')
    elif part[2] == 'ANG':
        file_emotion.append('Anger')
    elif part[2] == 'FEA':
        file_emotion.append('Fear')
    elif part[2] == 'HAP':
        file_emotion.append('Happiness')
    elif part[2] == 'NEU':
        file_emotion.append('Neutral')
    elif part[2] == 'DIS':
        file_emotion.append('Disgust')
    else:
        file_emotion.append('Unknown')


    file_id.append(int(part[0]))


filename_df = pd.DataFrame(file_name, columns=['Name'])
path_df = pd.DataFrame(file_path, columns=['Path'])
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])
speaker_id_df = pd.DataFrame(file_id, columns=['ActorID'])

Crema_df = pd.concat([filename_df,path_df,emotion_df, speaker_id_df], axis=1)
Crema_df['source'] = 'cremad'
cdataset = Crema_df
cemotions = ['Sadness', 'Happiness', 'Anger', 'Fear','Disgust', 'Neutral']
cdataset = cdataset[cdataset['Emotions'].isin(cemotions)].reset_index(drop = True)
cdataset['Emotions'].value_counts()
```

```python
from keras.layers import Input, Conv1D, Dense, concatenate, Flatten
from keras.models import Model
import tensorflow as tf

def create_model(input1 ):



    # Define the CNN layers for each modality
    x1 = Conv1D(32, 3, activation='relu', padding='same')(input1)
    x1 = tf.keras.layers.MaxPooling1D()(x1)
    x1 = Conv1D(64, 3, activation='relu', padding='same')(x1)
    x1 = tf.keras.layers.MaxPooling1D()(x1)
    x1 = Flatten()(x1)

#     x2 = Conv1D(32, 3, activation='relu', padding='same')(input_mfcc)
#     x2 = tf.keras.layers.MaxPooling1D()(x2)
#     x2 = Conv1D(64, 3, activation='relu', padding='same')(x2)
#     x2 = tf.keras.layers.MaxPooling1D()(x2)
#     x2 = Flatten()(x2)

#     x3 = Conv1D(32, 3, activation='relu', padding='same')(input_conformer)
#     x3 = tf.keras.layers.MaxPooling1D()(x3)
#     x3 = Conv1D(64, 3, activation='relu', padding='same')(x3)
#     x3 = tf.keras.layers.MaxPooling1D()(x3)
#     x3 = Flatten()(x3)

    # Concatenate the features from different modalities
#     concat = concatenate([x1, x2, x3], axis=1)

    # Define the shared dense layers
    x = Dense(256, activation='relu')(x1)
    x = tf.keras.layers.BatchNormalization()(x)
    x = Dense(64, activation='relu')(x)
    x = tf.keras.layers.Dropout(0.1)(x)
    x = Dense(56, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.005))(x)
    return x


def final_model(inp1):

    x = create_model(inp1)
    sid = Dense(91, activation='softmax', name='ActorID')(x)
    emo = Dense(6, activation='softmax', name='Emotions')(x)
    age =Dense(1, activation='linear', name='Age')(x)
    gender=Dense(2, activation='softmax', name='Sex')(x)
    model = Model(inputs=inp1, outputs=[sid, emo, age, gender])  # Specify multiple inputs as a list

    return model

input1 = Input(shape=[1024, 1])
# input_mfcc = Input(shape=[40, 1])
# input_conformer = Input(shape=[1024, 1])

# Create the model
# model = final_model(input_speakrec, input_mfcc, input_conformer)
model = final_model(input1)
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
lr = 1e-3
import tensorflow_addons as tfa
#optimizer = tfa.optimizers.RectifiedAdam(learning_rate= lr)
optimizer = tf.keras.optimizers.Adam(learning_rate= lr)
age_loss = tf.keras.losses.MeanSquaredError()  # Example loss function for age prediction
gender_loss = tf.keras.losses.SparseCategoricalCrossentropy()  # Example loss function for binary gender prediction
speaker_id_loss = tf.keras.losses.SparseCategoricalCrossentropy()  # Example loss function for speaker ID
emotion_loss = tf.keras.losses.SparseCategoricalCrossentropy()
# Compile the model with the Riemannian optimizer
model.compile(
    optimizer=optimizer,

    loss={
        'Age': age_loss,
        'Sex': gender_loss,
        'ActorID': speaker_id_loss,
        'Emotions': emotion_loss
    },#from_logits=True
    loss_weights={
        'Age': 1.0,  # Adjust these weights based on the importance of each task
        'Sex': 2.0,
        'ActorID':25.0,
        'Emotions': 3.0
    },
    metrics={
        'Age': [tf.keras.metrics.RootMeanSquaredError()],
        'Sex': 'accuracy',  # Assuming binary classification
        'ActorID': 'accuracy',  # Assuming categorical classification
        'Emotions': 'accuracy'  # Assuming categorical classification
    }  #[tf.keras.metrics.SparseCategoricalAccuracy()],
)


reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor = 'val_accuracy',
                                                 factor = 0.2,
                                                 patience = 1,
                                                 verbose = 1,
                                                 min_delta = 1e-4,
                                                 min_lr = 1e-15,
                                                 mode = 'max')

earlystopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_accuracy',
                                                 min_delta = 1e-4,
                                                 patience = 70,
                                                 mode = 'max',
                                                 restore_best_weights = True,
                                                 verbose = 1)

checkpointer = tf.keras.callbacks.ModelCheckpoint(filepath = './cnn_cremad.hdf5',
                                                  monitor = 'val_accuracy',
                                                  verbose = 1,
                                                  save_best_only = True,
                                                  save_weights_only = True,
                                                  mode = 'max')

callbacks = [earlystopping, checkpointer, reduce_lr]#reduce_lr]
```

```python
history = model.fit(
    x_train_conformer,
    {'ActorID': y_train_conformer['ActorID'], 'Emotions': y_train_conformer['Emotions'], 'Age': y_train_conformer['Age'], 'Sex': y_train_conformer['Sex']},
    validation_data=(x_val_conformer, {'ActorID': y_val_conformer['ActorID'], 'Emotions': y_val_conformer['Emotions'], 'Age': y_val_conformer['Age'], 'Sex': y_val_conformer['Sex']}),
    batch_size=32,
    epochs=150,
    callbacks=callbacks
)
```

```python
# Assuming you have test data: x_test, y_test
# x_test is your input test data, y_test contains all output labels

# Predict on test data
predictions = model.predict(x_val_conformer)

# Assuming predictions contain outputs for ['ActorID', 'Emotions', 'Age', 'Sex']
# Separate the predictions for each output
predictions_actor_id = predictions[0]
predictions_emotions = predictions[1]
predictions_age = predictions[2]
predictions_sex = predictions[3]

# Assuming y_test contains labels for all outputs in the same order
y_test_actor_id = y_val_conformer['ActorID']
y_test_emotions = y_val_conformer['Emotions']
y_test_age = y_val_conformer['Age']
y_test_sex = y_val_conformer['Sex']

# Import classification_report from sklearn.metrics
from sklearn.metrics import classification_report

# Generate classification reports for each output
report_actor_id = classification_report(y_test_actor_id, predictions_actor_id.argmax(axis=1),  digits=4)
report_emotions = classification_report(y_test_emotions, predictions_emotions.argmax(axis=1), digits=4)

report_sex = classification_report(y_test_sex, predictions_sex.argmax(axis=1),  digits=4)

# Print the classification reports
print("Classification Report for Actor ID:")
print(report_actor_id)

print("\nClassification Report for Emotions:")
print(report_emotions)


print("\nClassification Report for Sex:")
print(report_sex)
```
